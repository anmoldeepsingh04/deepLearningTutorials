## What is a Neuron?
Suppose we have the following task:
Given an age of a person, come up with a function that can predict if a person will buy insurance or not.

```python
# loading and visualizing the data
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('insurance_data.csv')

plt.scatter(df['age'], df['bought_insurance'], color = 'red', marker = '+')
plt.xlabel('Age')
plt.ylabel('Yes or No?')
plt.show()
```

```python
# implementing the linear regression model

from sklearn.linear_model import LinearRegression as LR
lr = LR()

# creating train-test split
from sklearn.model_selection import train_test_split as tts
X_train, X_test, y_train, y_test = tts(df[['age']].values, df['bought_insurance'], test_size = 0.2)

# training the model
lr.fit(X_train, y_train)

# testing the model
lr.score(X_test, y_test)
```

```python
# visualising the data and the model
plt.scatter(df['age'], df['bought_insurance'], color = 'red', marker = '+')
plt.plot(df['age'], lr.coef_[0]*df['age'] + lr.intercept_)
plt.xlabel('Age')
plt.ylabel('Yes or No?')
plt.show()
```

Next we provide the equation of the line as an input for the logit function and map the outputs to 0 and 1, which basically means if the person has taken health insurance or not.



We can collect the processes in a single cirlce and call it a neuron, where an input value is first put through a function and then through a activation function, and finally we get an output which is a value between 0 and 1. We can treat the logistic regression as a single neuron.



Here we just had one features (age), but we can have multiple features and correspondingly the equation would also change. The constants are called weights and the variables are called features.
