## Derivatives

Put very simply, the derivative for a function gives the slope of that function at that point.

For linear function, the derivate is the slpoe of the same line.



For a complex function, the slope is a function itself and changes as we change our point of interest.



Now for functions of multiple inputs, we define the partial derivative, which is the change in the the function value as one of the inputs is changed. We treat the oter input as a constant and differentiate the function with respect to the variable of interest.
The following function demonstrates the partial derivatives:
$$
f(x,y) = x^3 + y^2
$$

$$
\frac{\partial f}{\partial x} = 3x^2 + 0 = 3x^2
$$


$$
\frac{\partial f}{\partial y} = 0 + 2y = 2y

$$

Derivatives are needed when we are training our neural network and we want to modify the weights using back-propagation.



We differentiate the error with respect to each input, find the change in error when one of the inputs is changed and then adjust the weights accordingly.
