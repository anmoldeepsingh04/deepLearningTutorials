```python
import numpy as np
revenue = np.array([[100, 200, 220], [24, 36, 40], [12, 18, 20]])
expenses = np.array([[80, 90, 100], [10, 16, 20], [8, 10, 10]])

# matrix multiplication
profit = revenue - expenses
profit
```

```python
price_per_unit = np.array([1000, 400, 1200])
units = np.array([[30, 40, 50], [5, 10, 15], [2, 5, 7]])

# here the broadcasting concept is used, when the matrices aren't of the same size, then numpy changes the smaller matrix as if it was a 3x3 matrix with the first row repeated
# revenue = price_per_unit*units
# revenue

# to get the actual product, we use the dot product
revenue = np.dot( price_per_unit, units)
revenue
```
